```
import pandas as pd
pd.set_option('display.max_colwidth', -1) #to see more text

import numpy as np
import seaborn as sns

import itertools
from collections import Counter
```

```
from nltk.tokenize import TweetTokenizer
tknzr = TweetTokenizer()
```

```
# download and open some boomer tweets from trump_tweet_data_archive

boomer_tweets_df = pd.read_json('https://github.com/CALDISS-AAU/sdsphd19_coursematerials/raw/master/data/tweets_boomer.zip')
```

```
# Reset index (not really needed but why not)
boomer_tweets_df = boomer_tweets_df.set_index(pd.to_datetime(boomer_tweets_df.created_at))
boomer_tweets_df
```

```
# Let's identify people hashtag
boomer_tweets_df['rehashtags'] = boomer_tweets_df['tweet'].map(lambda textline: [tag for tag in tknzr.tokenize(textline) if tag.startswith('#')])
```

```
# Only keep tweets where a mention i present
boomer_tweets_df = boomer_tweets_df[boomer_tweets_df['rehashtags'].map(len) > 0]
```

```
# Collect
boomer_tags = itertools.chain(*boomer_tweets_df['rehashtags'])
taglist = list(set(boomer_tags))
```

```
taglist
```

```
counted_tags = Counter(taglist)
#counted_tags
counted_tags.most_common()[:]
#counted_tags.most_common()
```
